```python
from tuulUtils.facedetector import FaceDetector
import cv2
from matplotlib import pyplot as plt

img_path = 'images/putin.jpg'
face_cascade_path = 'haarcascades/haarcascade_frontalface_alt.xml'

image = cv2.imread(img_path)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
gray = cv2.cvtColor(image, cv2.COLOR_RGB2GRAY)

fd = FaceDetector(face_cascade_path)
faceRects = fd.detect(gray, scaleFactor=1.05, minNeighbors=5, minSize=(25,25))
print('I found {} faces'.format(len(faceRects)))

for (x, y, w, h) in faceRects:
    cv2.rectangle(image, (x,y), (x+w, y+h), (0, 255, 0), 2)
    cv2.putText(image, 'Face Detected', (x, y-10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (36,255,12), 2)


fig = plt.figure(figsize=(15,8))
ax1 = fig.add_subplot(1,1,1)
ax1.set(title='Detect Image')
ax1.imshow(image)
```
